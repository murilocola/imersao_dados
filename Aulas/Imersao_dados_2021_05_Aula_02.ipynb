```python


plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto',
              data=df.query('composto in @df_composto') ,
              order=df.query('composto in @df_composto')['composto'].value_counts().index
    )
ax.set_title('TOP 5')
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto',
              data=df.query('composto in @df_composto') ,
              order=df.query('composto in @df_composto')['composto'].value_counts().index,
              hue="composto"              
)
ax.set_title('TOP 5')
#ax.set(xlabel="Compostos", ylabel="Qtd")
ax.set_xlabel("Compostos",fontsize=20)
ax.set_ylabel("Qtd.",fontsize=20)
plt.show()
```
